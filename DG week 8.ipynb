```python
data = pd.read_csv('Test.csv')
```

```python
#look for outliers
data.info()   #look at data frame
data.describe()   #have a look at mean, medians and quartiles to see if the data is skewed or if there are any outliers

#from .describe() we can see that there appears to be an outlier in age as the maximum age is 164 which can't be right
```

```python
#checking for values that aren't supposed to be there, replace with P, passive

valid_values = ['N', 'F', 'P', 'B', 'A']
for i, value in enumerate(data['employee index']):
    if value not in valid_values:
        data.at[i, 'employee index'] = 'U'   #for unknown
        
valid_values = [1, 2, 3, 4, 'P']
for i, value in enumerate(data['indrel_1mes']):
    if value not in valid_values:
        data.at[i, 'indrel_1mes'] = 'U'   #for unknown
        
valid_values = ['S', 'N']
for i, value in enumerate(data['indresi']):
    if value not in valid_values:
        print(value)   #no missing data
        
valid_values = ['S', 'N']
for i, value in enumerate(data['indext']):
    if value not in valid_values:
        print(value)   #no missing data

valid_values = [1, 0]   #for spouse employee
for i, value in enumerate(data['conyuemp']):
    if value not in valid_values:
        data.at[i, 'conyuemp'] = 'U'  #to indicate unknown
```

```python
valid_values = ['V', 'H']
for i, value in enumerate(data['sex']):
    if value not in valid_values:
        data.at[i, 'sex'] = 'U'  #to indicate unknown
```

```python
valid_values = range(0,120)   #age range, set limit to 120 as likely no one older than that
for i, value in enumerate(data['age']):
    if value not in valid_values:
        data.at[i, 'indfall'] = 'N'   #mistakenly put as not deceased
        data.at[i, 'age'] = 'U'  #for unknown
```
